```python
# Lab 04212025

# Using PCA to reduce the dimensionality of a 2D dataset and visualizing the captured variance:

# 1):

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
X = np.random.multivariate_normal(mean=[2, 5], cov=[[3, 2], [2, 2]], size=100)

plt.scatter(X[:, 0], X[:, 1], alpha=0.7)
plt.title("Original 2D Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.axis("equal")
plt.grid(True)
plt.show()

# 2):

X_mean = np.mean(X, axis=0)
X_centered = X - X_mean

# 3):

cov_matrix = np.cov(X_centered.T)
print("\nCovariance Matrix:\n", cov_matrix)

# 4):

eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
print("\nEigenvalues:\n", eigenvalues)
print("\nEigenvectors:\n", eigenvectors)
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]

# 5):

pc1 = eigenvectors[:, 0]
X_pca_1D = X_centered @ pc1

# 6):

plt.scatter(X_pca_1D, np.zeros_like(X_pca_1D), alpha=0.7)
plt.title("1D Projection onto First Principal Component")
plt.xlabel("Principal Component 1")
plt.grid(True)
plt.show()

X_proj_2D = np.outer(X_pca_1D, pc1) + X_mean
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], alpha=0.5, label='Original Data', color='skyblue')
plt.scatter(X_proj_2D[:, 0], X_proj_2D[:, 1], alpha=0.8, label='1D Projection', color='green')
pc1_line = np.array([X_mean - 3 * pc1, X_mean + 3 * pc1])
plt.plot(pc1_line[:, 0], pc1_line[:, 1], color='red', linewidth=2, label='PC1 (Direction)')
plt.title("Comparison of 1D PCA Projection and  Original Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()

# 7):

explained_variance_ratio = eigenvalues / np.sum(eigenvalues)

plt.bar([1, 2], explained_variance_ratio * 100)
plt.title("Variance Explained by Each Principal Component")
plt.xlabel("Principal Component")
plt.ylabel("Variance Explained (%)")
plt.xticks([1, 2])
plt.grid(True)
plt.show()
```

```python
# Using t-SNE to visualize how it separates points from different classes in 2D space:

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
import warnings

warnings.filterwarnings("ignore", category=FutureWarning)

digits = load_digits()
X = digits.data      
y = digits.target   

# 1): 

print("Shape of X:", X.shape)
print("Shape of y: ({})".format(y.shape[0]))
print()
print("The 64 features represent that each sample is an 8x8 image, flattened into 64 features.")

# 2):

tsne = TSNE(n_components=2, random_state=42, perplexity=30)
X_tsne = tsne.fit_transform(X)

# 3): 

plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='tab10', alpha=0.7)
plt.colorbar(scatter, ticks=range(10))
plt.title("t-SNE Visualization (Perplexity = 30)")
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.grid(True)
plt.show()

# 4):

perplexities = [5, 30, 50]
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for i, perplexity in enumerate(perplexities):
    tsne = TSNE(n_components=2, perplexity=perplexity, random_state=42)
    X_embedded = tsne.fit_transform(X)
    scatter = axes[i].scatter(X_embedded[:, 0], X_embedded[:, 1], c=y, cmap='tab10', alpha=0.7)
    axes[i].set_title(f"Perplexity = {perplexity}")
    axes[i].set_xlabel("Component 1")
    axes[i].set_ylabel("Component 2")

plt.tight_layout()
plt.show()

print('Comparisions:')
print()
print('Perplexity = 5:')
print('Focuses more on local relationships.')
print('Clusters are more spread out and fragmented.')
print('Show tight mini-groups but less separation globally.')
print()
print('Perplexity = 30:')
print('good balance between local and global structure.')
print('Clusters are well-formed and separated.')
print('Maintains global structure while still showing clear groupings.')
print()
print('Perplexity = 50:')
print('Focuses more on global structure.') 
print('Clusters appear closer to each other, slightly less defined boundaries between some digits.')
print('Can be helpful for spotting higher-level structure, but may merge nearby groups.')
```
